```python
import re
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import matplotlib
```

```python
data = pd.read_csv('Age-standardized-suicide-rates.csv')
```

```python
A_data = data.loc[data['Location'] == 'United States of America']
```

```python
years = [2000,2002,2004,2006,2008,2010,2012,2014,2016,2018]
M_plot_A = A_data.loc[A_data['Dim1'] == 'Male'].plot('Period','FactValueNumeric',kind = 'line',c = 'b',label = 'Males')
F_plot_A = A_data.loc[A_data['Dim1'] == 'Female'].plot('Period','FactValueNumeric',kind = 'line',ax = M_plot_A,c = 'pink',label = 'Females')
plot_A = A_data.loc[A_data['Dim1'] == 'Both sexes'].plot('Period','FactValueNumeric',kind = 'line',ax = F_plot_A,c = 'black',label = 'Both sexes',xticks = years)
plt.title('Aveage Suicides Over Time in the United States')
plt.xlabel('Year')
plt.ylabel('Suicides per 100,000')
plt.show()
```

```python
def MakePlots(df,country):
    df = df.loc[df['Location'] == country]
    M_plot = df.loc[df['Dim1'] == 'Male'].plot('Period','FactValueNumeric',kind = 'line',c = 'b',label = 'Males')
    F_plot = df.loc[df['Dim1'] == 'Female'].plot('Period','FactValueNumeric',kind = 'line',ax = M_plot,c = 'pink',label = 'Females')
    plot = df.loc[df['Dim1'] == 'Both sexes'].plot('Period','FactValueNumeric',kind = 'line',ax = F_plot,c = 'black',label = 'Both sexes',xticks = years)
    plt.title('Aveage Suicides Over Time in ' + country)
    plt.xlabel('Year')
    plt.ylabel('Suicides per 100,000')
    plt.show()
MakePlots(data,'Rwanda')
```
